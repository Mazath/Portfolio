```python
#imports
import requests
from bs4 import BeautifulSoup
import csv
```

```python
#set URL to scrape

targetURL = 'https://en.wikipedia.org/wiki/List_of_universities_in_Australia'
```

```python
#request html from target

html = requests.get(targetURL)

#output status code to ensure request worked. if successful, output will print 200
print(html.status_code)
```

```python
#create BS object with html

soup = BeautifulSoup(html.text)

#output the soup object to ensure it is populated
print(soup.prettify())
```

```python
#create variable with the target table html

table = soup.find('table', class_ = 'wikitable sortable')

#create variable with all table rows

rows = table.find_all('tr')
```

```python
#create variable with the header row html, removing return characters

#test we can add.strip()

columns = [v.text.replace('\n', ' ') for v in rows[0].find_all('th')]

print (columns)
```

```python
#create variable with the column header rows for export to the CSV file

headers = [str(columns[0]), str(columns[1]), str(columns[2]) ,str(columns[3]),
 str(columns[4]), str(columns[5])]

print(headers)

```

```python
#remove 'unwanted text from the campus and state/territory column headers

headers[2] = headers[2][:-4] 

headers[3] = headers[3][:-4] 

headers

print (headers)
```

```python
#create variable to hold rows for export 

holder_rows = []

#iterate through rows an obtain target text
for i in range(1,len(rows)):
    tds = rows[i].find_all('td')

    if len(tds) == 11:

        #strips any return character from text and appends to temp variable 'values'
        values = [tds[0].text.replace('\n', ''), 
        tds[1].text.replace('\n', ''), 
        tds[2].text.replace('\n', ''), 
        tds[3].text.replace('\n', ''), 
        tds[4].text.replace('\n', ''),
        tds[5].text.replace('\n', '')]
        #appends row values to the holder_rows object
        holder_rows.append(values)

```

```python
#print holder_rows to check it's populated

print(holder_rows)
```

```python
#export data to to csv

#open the export file in output directory
with open('../output/uniTable.csv', 'w', newline= '') as f:
    #create the writer object
    CSV_writer = csv.writer(f)

    #write the headers
    CSV_writer.writerow(headers)
    #write thr rows
    CSV_writer.writerows(holder_rows)

#close the file
f.close()
```
